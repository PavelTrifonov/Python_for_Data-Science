### 1. В чём разница между зависимыми и независимыми выборками?

### Зависимые выборки:

*Зависимые выборки* (или связанные выборки) — это выборки, где элементы одной выборки непосредственно связаны или попарно сопоставлены с элементами другой выборки.

**Примеры:**

- Повторные измерения: Исследование, где одно и то же измерение проводится на одной и той же группе людей в разное время (например, до и после лечения).
- Парные сравнения: Исследование, где каждый участник экспериментальной группы имеет свой аналог в контрольной группе по определённым параметрам (например, возраст, пол, вес).

**Особенности:**

- Обычно анализируют разницу между парными значениями.
- При меняют специальные статистические методы, такие как парный t-тест.

### Независимые выборки:

*Независимые* выборки - это выборки, где элементы одной выборки не связаны с элементами другой выборки.

**Примеры:**

- Разные группы: Сравнение двух разных групп людей, где одна группа получала лечение, а другая - нет.
- Разные условия: Сравнение результатов двух разных методов лечения на разных группах людей.

**Особенности:**
- Сравнение проводится между средними или другими двух независимыми группами.
- При меняют такие статистические методы, как независимый t-тест.

### 2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

Параметрические и непараметрические статистические критерии касаются зависимости от свойств данных и предположений, которые можно сделать о распределении этих данных.

### Параметрические критерии:

- Нормальность распределения: Данные должны иметь нормальное распределение. Это один из основных предпосылок.
- Гомогенность дисперсий: Дисперсия в сравниваемых группах должна быть равной (гомоскедастичность).
- Интервальные и шкалы отношений: Данные должны быть измерены в интервальной или шкале отношений.
- Большие выборы: Параметрические тесты более мощные и психологические, особенно на больших выборах.

**Примеры**:

- t-тест для независимых выборов.
- Парный t-тест.
- ANOVA (дисперсионный анализ).
- Пирсоновская корреляция.

### Непараметрические критерии:

- Отсутствие нормальности: Данные не распределены нормально или не подлежат проверке нормальности.
- Малые выборы: Выборка Малайзии, и сложно решить предпосылку параметрических конфликтов.
- Непрерывные или порядковые шкалы: Данные могут быть измерены в порядковой шкале (ранговые данные), или имеются выбросы, влияющие на параметры распределения.
- Неравенство дисперсий: Дисперсии в группах могут быть неравны (гетеросклерозность).

**Примеры**:

- U-критерий Манна-Уитни (аналог t-теста для независимых выборов).
- Критерий Уилкоксона (аналог парного t-теста).
- Критерий Краскела-Уоллиса (аналог однофакторного ANOVA).
- Спирменовская ранговая корреляция (аналог Пирсоновской корреляции).
