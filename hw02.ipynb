**Задание 3**

- С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3
включительно.
- Создайте массивы y1, y2, y3, y4 по следующим формулам:
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = sin(x)
- Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с
названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных
графика в сетке, состоящей из двух строк и двух столбцов. 
- В каждом графике массив x используется
для координат по горизонтали.
- В левом верхнем графике для координат по вертикали используйте
y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4. 
- Дайте название графикам:
'График y1', 'График y2' и т.д.
- Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
- Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
- Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.

```python
x = np.linspace(-3,3,51)
x
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig,ax = plt.subplots(2,2,figsize=(12,8)) # вместо 8 и 6 установили 12 и 8 дюймов
ax1,ax2,ax3,ax4 = ax.flatten()

ax1.set_title('График y1')
ax2.set_title('График y2')
ax3.set_title('График y3')
ax4.set_title('График y4')

ax1.plot(x, y1, linewidth=3)
ax1.set_xlim([-5, 5])
ax2.plot(x, y2, label='2 * x + 0.5', color='green', linewidth=3)
ax3.plot(x, y3, color='orange',linewidth=3)
ax4.plot(x, y4, color='red')

fig.subplots_adjust(hspace=0.3, wspace=0.3)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
creditcard_df = pd.read_csv('creditcard.csv')
df_Class = creditcard_df.Class.value_counts()
df_Class
```

```python
df_Class.plot(kind="bar",logy=False,figsize=(6,8))
```

```python
df_Class.plot(kind="bar",logy=True,figsize=(6,8))
```

```python
class_1 = creditcard_df[creditcard_df.Class == 1].V1
class_1
```

```python
class_0 = creditcard_df[creditcard_df.Class == 0].V1
class_0
```

```python
fig,ax = plt.subplots(1,2,figsize=(8,6))
fig.subplots_adjust(hspace=0.3, wspace=0.7)

ax_1,ax_0 = ax.flatten()
class_1.hist(ax=ax_1,density=True,bins=30)
ax_1.set_xlabel('value')
ax_1.set_ylabel('V1')

class_0.hist(ax=ax_0,density=True,bins=30)
ax_0.set_xlabel('value')
ax_0.set_ylabel('V1')

ax_1.set_title('Class_1')
ax_0.set_title('Class_0')
plt.show()
```

```python
plt.plot(figsize=(7,5))

class_1.hist(density=True,bins=20,alpha=0.5,color='r')
class_0.hist(density=True,bins=20,alpha=0.5,color='grey')
plt.legend(['class_1','class_0'])
plt.xlabel('V1')

plt.show()
```
