### Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

В задачах машинного обучения и обработки данных разные метрики качества используются в зависимости от цели и контекста применения. Precision и Recall являются важными метриками, которые помогают оценивать качество классификационных моделей. В зависимости от задачи можно решать, какую из этих метрик следует максимизировать.

**Когда лучше максимизировать Precision**

Precision (точность) – это доля правильно предсказанных положительных наблюдений из всех наблюдений, которые модель предсказала как положительные. Precision важен в тех случаях, когда ложноположительные результаты могут быть дорогостоящими или критичными.

**Примеры:**

- *Медицинская диагностика:* В случае редких заболеваний важно минимизировать количество ложноположительных диагнозов, чтобы не подвергать пациентов ненужным и потенциально опасным процедурам или лечению.
- *Финансовые транзакции:* При обнаружении мошенничества важно, чтобы каждая отмеченная транзакция как мошенническая действительно была мошеннической, чтобы не беспокоить клиентов по ошибке и не ухудшать их клиентский опыт.
- *Спам-фильтры:* Важно, чтобы как можно меньше легитимных писем попало в папку со спамом, чтобы пользователи не пропускали важные сообщения.

**Когда лучше максимизировать Recall**

Recall (полнота) – это доля правильно предсказанных положительных наблюдений из всех истинных положительных наблюдений. Recall важен в тех случаях, когда важно найти все положительные случаи, даже если это приведет к большему числу ложноположительных результатов.

**Примеры:**

- *Обнаружение заболеваний:* Для ранней диагностики заболеваний (например, рака) важно найти как можно больше случаев болезни, даже если это приведет к некоторым ложноположительным результатам, чтобы начать лечение как можно раньше.
- *Поиск и спасение:* В задачах поиска и спасения людей важно максимизировать Recall, чтобы не пропустить ни одного человека, находящегося в опасности.
- *Системы безопасности:* В системах видеонаблюдения для обнаружения вторжений или угроз важно обнаружить все потенциальные угрозы, даже если это приведет к некоторым ложным срабатываниям.



# Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

В большинстве реальных задач необходимо найти баланс между Precision и Recall, который будет оптимальным для конкретного контекста. Для этого часто используют такие метрики, как F1-score, которая является гармоническим средним между Precision и Recall и помогает найти компромисс между этими двумя метриками.

*F1-score:* Используется, когда важно как правильно предсказать положительные случаи, так и найти все возможные положительные случаи.
